### Simple Linear Regression w/ Feature Engineering

#### Absolute Price

```python
X = df.drop(columns=['log1p_price','Unnamed: 0', 'price', 'artist', 'title', 'size', 'url', 'painting type', 'path', 'material', 'style'])
y = df['price']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = np.abs(y_test-y_pred).mean()
r2 = model.score(X_test, y_test)

print(f"rmse: {round(rmse,2)}, mae: {round(mae,2)}, r2: {round(r2,2)}")
```

#### Log price

```python
X_log = df.drop(columns=['log1p_price', 'Unnamed: 0', 'price', 'artist', 'title', 'size', 'url', 'painting type', 'path', 'material', 'style'])
y_log = df['log1p_price']

X_log_train, X_log_test, y_log_train, y_log_test = train_test_split(
    X_log, y_log, test_size=0.3, random_state=42
)

model = LinearRegression()
model.fit(X_log_train, y_log_train)

y_log_pred = model.predict(X_log_test)
rmse_log = np.sqrt(mean_squared_error(y_log_test, y_log_pred))
mae_log = np.abs(y_log_test-y_log_pred).mean()
r2_log = model.score(X_log_test, y_log_test)

print(f"rmse: {round(rmse_log,2)}, mae: {round(mae_log,2)}, r2: {round(r2_log,2)}")

# Translating to Absolute Price
y_log_test
```

#### Checking correlations and VIF

```python
corr_matrix = X.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```

```python
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]

print("\nFator de Inflação de Variância (VIF):")
print(vif_data)
```

```python
X_new = X
X_new = X_new.drop(columns=[i for i in vif_data[vif_data['VIF'] > 5]['feature']])
X_new.columns
y_new = df['price']

corr_matrix = X_new.corr()
plt.figure(figsize=(15, 15))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', center=0)
plt.title('Matriz de Correlação entre Features')
plt.show()
```

#### Training in Optimal features (VIF<5) - Normal and Log prices

```python
X_new_train, X_new_test, y_new_train, y_new_test = train_test_split(
    X_new, y_new, test_size=0.3, random_state=42
)

model_new = LinearRegression()
model_new.fit(X_new_train, y_new_train)

y_new_pred = model_new.predict(X_new_test)
rmse_new = np.sqrt(mean_squared_error(y_new_test, y_new_pred))
mae_new = np.abs(y_new_test-y_new_pred).mean()
r2_new = model_new.score(X_new_test, y_new_test)

print(f"rmse: {round(rmse_new,2)}, mae: {round(mae_new,2)}, r2: {round(r2_new,2)}")

X_new_train_log, X_new_test_log, y_new_train_log, y_new_test_log = train_test_split(
    X_new, y_log, test_size=0.3, random_state=42
)

model_new_log = LinearRegression()
model_new_log.fit(X_new_train_log, y_new_train_log)

y_new_pred_log = model_new_log.predict(X_new_test_log)
rmse_log = np.sqrt(mean_squared_error(y_new_test_log, y_new_pred_log))
mae_log = np.abs(y_new_test_log-y_new_pred_log).mean()
r2_log = model_new_log.score(X_new_test_log, y_new_test_log)

print(f"rmse: {round(rmse_log,2)}, mae: {round(mae_log,2)}, r2: {round(r2_log,2)}")
```

### Other models

#### Training on Absolute Price

```python
# Define a dictionary of models you want to train
other_models = {
    'Linear Regression': LinearRegression(),
    'SGD Regressor': SGDRegressor(),
    'Ridge': Ridge(),
    'Lasso': Lasso(),
    'Elastic Net': ElasticNet(),
    'Random Forest Regressor': RandomForestRegressor(),
    'XGBoost Regressor': XGBRegressor(),
    'LightGBM Regressor': LGBMRegressor(),
    'CatBoost Regressor': CatBoostRegressor(verbose=0) # Added verbose=0 to avoid excessive output
}

# Assuming X_new_train, X_new_test, y_new_train, y_new_test are already defined
# from your previous data splitting step with VIF filtered features

results = {}

# Iterate over the models
for model_name, model in other_models.items():
    print(f"Training {model_name}...")

    # Train the model
    model.fit(X_new_train, y_new_train)

    # Make predictions
    y_pred = model.predict(X_new_test)

    # Calculate evaluation metrics
    rmse = np.sqrt(mean_squared_error(y_new_test, y_pred))
    mae = np.abs(y_new_test - y_pred).mean()
    r2 = model.score(X_new_test, y_new_test)

    # Store the results
    results[model_name] = {
        'RMSE': round(rmse, 2),
        'MAE': round(mae, 2),
        'R2': round(r2, 2)
    }

# Print the results for each model
print("\nModel Evaluation Results:")
for model_name, metrics in results.items():
    print(f"{model_name}:")
    for metric_name, value in metrics.items():
        print(f"  {metric_name}: {value}")
```
